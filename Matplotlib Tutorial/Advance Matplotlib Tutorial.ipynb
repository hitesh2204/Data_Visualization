### Importing Library.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("sharma-kohli.csv")
```

```python
df
```

### Plotting 2-D line chart.
1) It used to perform Univariate/Bivariate analysis.

2) It is used to perform on numerical columns and Categorical data.

3) It mostly work on TimeSeries data.

```python
plt.plot(df['V Kohli'],df['index'],color='red',marker='*',linewidth=5)
plt.title("Virat Kohli Record")
plt.xlabel("Kohli Runs")
plt.ylabel("Year")
plt.savefig("kohli record")
plt.show()
```

```python
plt.plot(df['V Kohli'],df['index'],color='red',marker='*',linewidth=5,label='Kohli')
plt.plot(df['RG Sharma'],df['index'],color='green',marker='*',linewidth=5,label='Rohit')
plt.title("Virat Kohli vs Rohit Sharma Record")
plt.xlabel("Kohli/Rohit Runs")
plt.ylabel("Year")
plt.legend(loc='best')
plt.savefig("kohli record")
plt.show()
```

### Scatter Plot.
The scatter plot is used on numerical data.

It's used in bivariate analysis.

```python
plt.scatter(df['V Kohli'],df['RG Sharma'],c=df['index'],cmap='summer',marker='*',s=100)
plt.title("Vk and Rohit Runs")
plt.xlabel("Vk")
plt.ylabel("Rohit")
plt.colorbar()
plt.show()
```

### Bar Chart.

```python
plt.bar(df['V Kohli'],df['index'],width=10,color='blue')
plt.title("Vk Run as per Year")
plt.xlabel("Vk")
plt.ylabel("Runs")
plt.show()
```

#### Annotation.

```python
bat_df=pd.read_csv("batter.csv")
```

```python
bat_df=bat_df.head(100).sample(25,random_state=5)
```

```python
bat_df.head()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(bat_df['avg'],bat_df['strike_rate'],s=bat_df['runs'])
```

```python
### Annotate each dot.represent each dot as each batsman.
### text() function used to give name.
plt.figure(figsize=(12,8))
plt.scatter(bat_df['avg'],bat_df['strike_rate'],s=bat_df['runs'],marker='*')

for i in range(bat_df.shape[0]):
    plt.text(bat_df['avg'].values[i],bat_df['strike_rate'].values[i],bat_df['batter'].values[i])
```

### Plotting Horizontal and Vertical Lines.

```python
plt.figure(figsize=(12,8))
plt.scatter(bat_df['avg'],bat_df['strike_rate'],s=bat_df['runs'],marker='*')

plt.axhline(125,color='red')
plt.axhline(140,color='Blue')
plt.axvline(30,color='red')

for i in range(bat_df.shape[0]):
    plt.text(bat_df['avg'].values[i],bat_df['strike_rate'].values[i],bat_df['batter'].values[i],fontdict={'size':12,'color':'green'})
```

### Subplot Graph :-

```python
plt.subplots()

### It gives you two objects
### 1) One is figure level
### 2) Axes level
```

```python
df=pd.read_csv("batter.csv")
```

```python
df.head()
```

```python
### Plotting the normal scatter graph by using subplot objects.
fig,ax=plt.subplots()
ax.scatter(df['runs'],df['avg'])
ax.set_title("Batsman runs records",fontdict={'size':12})
ax.set_xlabel('Runs',fontdict={'size':12})
ax.set_ylabel('AVG',fontdict={'size':12})
fig.show()
```

```python
### Adding some additional information.
fig,ax=plt.subplots(figsize=(10,6))
ax.scatter(df['runs'],df['avg'],c=df['runs'],marker='*',s=df['runs'],label='Runs')
ax.set_title("Batsman runs records",fontdict={'size':12})
ax.set_xlabel('Runs',fontdict={'size':12})
ax.set_ylabel('AVG',fontdict={'size':12})
ax.legend()
fig.show()
```

### *** Plotting multiple graph at one time.

```python
df.head()
```

```python
fing,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,7),sharex=True)

ax[0].scatter(df['runs'],df['avg'],marker='*',c=df['runs'],s=df['runs'])
ax[1].scatter(df['runs'],df['strike_rate'],marker='+',c=df['runs'],s=df['runs'])

ax[0].set_title("Batsman Record")
ax[0].set_xlabel('runs',fontdict={'size':12})
ax[0].set_ylabel('avg',fontdict={'size':12})

ax[1].set_title("Batsman Record")
ax[1].set_xlabel('runs',fontdict={'size':12})
ax[1].set_ylabel('strike_rate',fontdict={'size':12})
```

```python
tip=pd.read_csv('tips.csv')
tip.head()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,7))

ax[0,0].scatter(tip['total_bill'],tip['tip'],color='green',marker='*')
ax[0,1].bar(tip['smoker'],tip['total_bill'],color='orange')
ax[1,0].hist(tip['total_bill'],color='blue')
ax[1,1].plot(tip['total_bill'],tip['size'])

ax[0,0].set_title("Hotel Record")
ax[0,0].set_xlabel('Total_bills',fontdict={'size':12})
ax[0,0].set_ylabel('Tips',fontdict={'size':12})

ax[0,1].set_title("Hotel Record")
ax[0,1].set_xlabel('Smoker',fontdict={'size':12})
ax[0,1].set_ylabel('Total_bill',fontdict={'size':12})

ax[1,0].set_xlabel('Range',fontdict={'size':12})
ax[1,0].set_ylabel('Total_bill',fontdict={'size':12})

ax[1,1].set_xlabel('Total_bill',fontdict={'size':12})
ax[1,1].set_ylabel('Tips',fontdict={'size':12})

fig.show()
```

### 3D surface plot:- 

```python
import numpy as np

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2 + yy**2
```

```python
#print(z)

fig=plt.figure(figsize=(8,6))

ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)
fig.show()
```

### Contour Graph:-

```python
fig=plt.figure(figsize=(8,6))

ax=plt.subplot()
ax.contour(xx,yy,z)
fig.show()
```

```python
fig=plt.figure(figsize=(8,6))

ax=plt.subplot()
ax.contourf(xx,yy,z)
fig.show()
```
