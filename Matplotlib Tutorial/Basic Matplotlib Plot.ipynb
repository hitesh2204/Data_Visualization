```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

### Type of Data.

### Types of Analysis
1) Univariate Analysis :- When we working on single column thats called as Univariate analysis.

2) Bivariate Analysis :- When we are working on two column thats called bivariate analysis.

3) Multivariate Analysis :- Whene we are working on more tha two columns thats called as multvariate analysis.

### 1) 2D-LINE PLOT :-

1) It used in Univariate / bivariate analysis.

2) It worked in Numerical and Categorical columns.

3) But it worked well when our one column is Numerical and other is Categorical.

4) Used on TImeSeries data.



```python
price=[47000,54000,40000,55000,35000,60000]
year=[2015,2017,2016,2020,2018,2019]
print("Price",price)
print("year",year)
```

```python
### Plotting  2d-line graph.
plt.plot(price,year)
plt.title("Test")
plt.xlabel("price")
plt.ylabel("year")
```

```python
crick_df=pd.read_csv("sharma-kohli.csv")
crick_df.head()
```

```python
### Plotting Season wise 2d-line graph between rohit and kohli.
plt.plot(crick_df['index'],crick_df['V Kohli'])
plt.title("Virat Kohli IPL-record")
plt.xlabel("Year",fontdict={'size':12})
plt.ylabel("Runs",fontdict={'size':12})
plt.grid()
plt.show()
```

```python
### Plotting vk and rohit ipl record at a time.
plt.plot(crick_df['index'],crick_df['V Kohli'],label='Kohli',color='red',marker='*')
plt.plot(crick_df['index'],crick_df['RG Sharma'],label='Rohit',color='green',marker='o')
plt.title("Virat and Rohit IPL-record")
plt.xlabel("Year",fontdict={'size':12})
plt.ylabel("Runs",fontdict={'size':12})
plt.legend()
plt.grid()
plt.show()
```

```python
df=pd.read_csv("batter.csv")
```

```python
new_df=df.head(10)
```

```python
new_df
```

```python
### 2-D Line graph on Numerical and Categorical columns.
plt.plot(new_df['batter'],new_df['avg'],color='green',marker='*',linewidth=3,markersize=10)
plt.title("Batsman Avg",fontdict={'size':12})
plt.xlabel("Batsman",fontdict={'size':12})
plt.ylabel("AVG")
plt.xticks(rotation=60)
plt.grid()
plt.show()
```

```python
### 2-D line plot on batsman and their strike rate.
plt.plot(new_df['batter'],new_df['strike_rate'],linewidth=3,marker='*',markersize=10,color='red',label='Batsman')
plt.title("Batsman and strik_rate")
plt.xlabel("Batsman")
plt.ylabel("strike_rates")
plt.xticks(rotation=60)
plt.legend()
plt.show()
```

```python
## Plotting 2_D graph on batsman and their strike_rate and runs.
plt.figure(figsize=(7,5))
plt.plot(new_df['batter'],new_df['strike_rate'],label='Strike_rate',color='green',linewidth=3,marker='*',markersize=10)
plt.plot(new_df['batter'],new_df['runs'],label='Runs',color='blue',linewidth=3,marker='o',markersize=10)
plt.title('batsman Runs vs Strike_rate',fontdict={'size':12})
plt.xlabel('batsman',fontdict={'size':12})
plt.ylabel('Runs and Strike_rate',fontdict={'size':12})
plt.xticks(rotation=60)
plt.grid()
plt.legend()
plt.show()
```

### 2) SCATTER-PLOT

1) It used in bivariate analysis.

2) It worked in Numerical -Numerical columns.

3) It is used for finding the corelation between two variable.

```python
batter=pd.read_csv("batter.csv")
batter.head()
```

```python
### Plotting the scatter on avg and strike_rate.

df=batter.head(100).sample(25,random_state=5)
df.shape
```

```python
plt.scatter(df['avg'],df['strike_rate'],marker='*',s=500)
plt.title("IPL analysis")
plt.xlabel("avg",fontdict={'size':12})
plt.ylabel("strike rate",fontdict={'size':12})
plt.show()
```

#### Increasing the size of data.

```python
plt.figure(figsize=(10,5))
plt.scatter(df['avg'],df['strike_rate'],marker='*',s=df['runs'])
plt.title("IPL analysis")
plt.xlabel("avg",fontdict={'size':12})
plt.ylabel("strike rate",fontdict={'size':12})
plt.show()
```

#### Showing both classes on different color.

```python
### Showing both classes on different color.
plt.figure(figsize=(10,5))
plt.scatter(df['avg'],df['strike_rate'],marker='*',s=df['runs'],c=df['runs'])
plt.title("IPL analysis")
plt.xlabel("avg",fontdict={'size':12})
plt.ylabel("strike rate",fontdict={'size':12})
plt.colorbar()
plt.show()
```

```python
iris=pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
iris.sample(5)
```

```python
### Potting the numerical data into scatter plot.
plt.figure(figsize=(10,6))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],marker='o',s=400,c=iris['Species'],cmap='viridis')
plt.title("Iris Data",fontdict={'size':12})
plt.xlabel("SepalLengthCm",fontdict={'size':12})
plt.ylabel('PetalLengthCm',fontdict={'size':12})
plt.colorbar()
for i in range(iris.shape[0]):
    plt.text(iris['SepalLengthCm'].values[i],iris['PetalLengthCm'].values[i],iris['Species'].values[i])
plt.show()
```

### 3) BAR-PLOT.

1) It used in bivariate analysis.

2) It worked in Numerical and categorical columns.

3) Numerical columns may be an aggregate function.

4) Used on aggregate analysis on group.

```python
color=['red','green','pink','black','brown','orange']
age=[10,15,12,18,20,19]
```

```python
plt.bar(color,age,color='green')
plt.title("Children data")
plt.xlabel("Color",fontdict={'size':12})
plt.ylabel("age",fontdict={'size':12})
plt.show()
```

```python
### Plotting the horizonral bar graph.

plt.barh(color,age,color='green')
plt.title("Children data")
plt.xlabel("Color",fontdict={'size':12})
plt.ylabel("age",fontdict={'size':12})
plt.show()
```

```python
plt.bar(color,age,color=['red','green','pink','blue','brown','orange'])
plt.title("Children data")
plt.xlabel("Color",fontdict={'size':12})
plt.ylabel("age",fontdict={'size':12})
plt.xticks(rotation=60)
plt.show()
```
