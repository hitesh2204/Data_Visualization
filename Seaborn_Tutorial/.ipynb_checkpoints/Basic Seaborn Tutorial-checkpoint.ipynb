### 1) Relational Plot :-

1) It is used to find out the statastical relationship between two or more variable.

2) It is used for Bivariate analysis.

3) Scatter and Line Plots are used.

#### 1) Scatter plot:-

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tip=pd.read_csv("tips.csv")
```

```python
tip.head()
```

```python
### This scatter plot is axis level function.
sns.scatterplot(data=tip,x='total_bill',y='tip')
```

```python
### We can plot scatter plot on figure level by useing Relplot.

sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',style='time',hue='sex')
```

```python
### Hue parameter.It will didvide the data on perticular class.
plt.figure(figsize=(8,5))
sns.scatterplot(data=tip,x='total_bill',y='tip',hue='sex',style='time',size='size',)
```

```python
gap=pd.read_csv("gapminder.csv")
```

```python
gap.head()
```

```python
Ind=gap[gap['country']=='India']
```

```python
sns.lineplot(data=Ind,x='year',y='life_exp')
```

```python
### Using relplot we can plot Line graph.

sns.relplot(data=Ind,x='year',y='life_exp',kind='line')
```

```python
## hue in line plot

countries=gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
countries.head()
```

```python
sns.relplot(data=countries,x='year',y='life_exp',kind='line',hue='country')
```

```python
sns.relplot(data=countries,x='year',y='life_exp',kind='line',hue='country',style='continent',size='continent')
```

### Facetplot:-
We can plot mulpltiple plot on categorical columns.

It only worked with figure level function.It is not used with scatter plot and line plot function.

```python
tip.head()
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',col='sex',kind='scatter',)
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',col='sex',row='smoker',kind='scatter')
```

```python
gap.head()
```

```python
plt.figure(figsize=(5,7))
sns.relplot(data=gap,x='life_exp',y='gdp_cap',kind='scatter',col='year',col_wrap=3)
```

### Distribution Plot:-
1) It is used for Univariate analysis

2) It is used for distribution

3) Range of observations

4) Central tendency

#### Fig level function :-> displot
#### axes level function :-> distplot , kde plot , rug plot

```python
### Plotting Univariatte Hiostogram.

tip.head()
```

```python
sns.histplot(data=tip,x='total_bill')
```

```python
sns.displot(data=tip,x='total_bill',kind='hist')
```

```python
### plotting histogram on categorical data.

sns.histplot(data=tip,x='day')
```

```python
### hue parameter.
sns.histplot(data=tip,x='tip',hue='sex')
```

```python
### Facet graph on histogram.
sns.displot(data=tip,x='day',col='sex',kind='hist')
```

### KDE plot :-

```python
tip.head()
```

```python
sns.kdeplot(data=tip,x='total_bill')
```

```python
sns.displot(data=tip,x='total_bill',kind='kde')
```

```python
sns.displot(data=tip,x='total_bill',kind='kde',hue='sex')
```

### Bivariate Histogram :-

```python
sns.histplot(data=tip,x='total_bill',y='tip')
```

### Bivariate KDE plot :-

```python
sns.kdeplot(data=tip,x='total_bill',y='tip')
```
